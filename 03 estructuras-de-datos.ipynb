## üìã Listas

### Concepto

Una **lista** es una colecci√≥n **ordenada** y **mutable** de elementos. Puede contener cualquier tipo de dato (n√∫meros, textos, booleanos, incluso otras listas).

- Se crean con **corchetes** `[]`.
- Los elementos se separan por **comas**.
- Cada elemento tiene una **posici√≥n (√≠ndice)** que empieza en **0**.

### Crear una lista

````python
# Lista de frutas
# frutas=[0=manzana,1=platano,2=fresa,3=uva]
frutas = ["manzana", "pl√°tano", "fresa", "uva"]
print(frutas)
````

````python
# Lista de n√∫meros
numeros = [10, 20, 30, 40, 50]
print(numeros)
````

````python
# Lista mixta
mixta = ["Ana", 25, True, 3.14]
print(mixta)
````

````python
# Lista vac√≠a
vacia = []
print(vacia)
````

### Acceder a elementos por √≠ndice

Cada elemento tiene una posici√≥n num√©rica llamada **√≠ndice**. El primer elemento tiene √≠ndice **0**.

| √çndice | 0 | 1 | 2 | 3 |
| --- | --- | --- | --- | --- |
| Elemento | "manzana" | "pl√°tano" | "fresa" | "uva" |

Tambi√©n puedes usar **√≠ndices negativos**: `-1` es el √∫ltimo, `-2` el pen√∫ltimo, etc.

````python
frutas = ["manzana", "pl√°tano", "fresa", "uva"]

print(frutas[0])   # manzana (primer elemento)
print(frutas[2])   # fresa (tercer elemento)
print(frutas[-1])  # uva (√∫ltimo elemento)
print(frutas[-2])  # fresa (pen√∫ltimo elemento)
````

### M√©todos principales de listas

Los **m√©todos** son funciones que pertenecen a un objeto. Las listas tienen muchos m√©todos √∫tiles.

#### `append()` ‚Äî Agregar un elemento al final

Agrega **un elemento** al **final** de la lista.

````python
compras = ["leche", "pan"]
print(compras)
compras.append("azucar")
print(compras)
````

#### `remove()` ‚Äî Eliminar un elemento por su valor

Elimina la **primera aparici√≥n** del valor indicado. Si no existe, da error.

````python
compras = ["leche", "pan", "huevos", "pan"]
print("Antes:", compras)

compras.remove("pan")  
print("Despu√©s:", compras)
````

#### `sort()` ‚Äî Ordenar la lista

Ordena los elementos **de menor a mayor** (por defecto). Con `reverse=True` ordena de mayor a menor.

‚ö†Ô∏è Solo funciona si todos los elementos son del **mismo tipo** (todos n√∫meros o todos textos).

````python
numeros = [42, 7, 99, 1, 23]
print("Original:", numeros)
numeros.sort()
print("Ordenada:", numeros)

numeros.sort(reverse=True)
print("Ordenada (descendente):", numeros)
````

#### `len()` ‚Äî Saber cu√°ntos elementos tiene la lista

`len()` no es un m√©todo de la lista, es una **funci√≥n integrada** de Python. Devuelve la **cantidad de elementos**.

````python
frutas = ["manzana", "pl√°tano", "fresa", "uva"]
total_elementos=len(frutas)
print("N√∫mero de frutas:", total_elementos)
````

#### `index()` ‚Äî Encontrar la posici√≥n de un elemento

Devuelve el **√≠ndice (posici√≥n)** de la primera aparici√≥n del valor buscado. Si no existe, da error.

````python
frutas = ["manzana", "pl√°tano", "fresa", "uva"]
posicion=frutas.index("fresa")
print("Posici√≥n de 'fresa':", posicion)
````

## üì¶ Tuplas

### Concepto

Una **tupla** es una colecci√≥n **ordenada** e **inmutable** de elementos.

- Se crean con **par√©ntesis** `()` (o simplemente separando con comas).
- **No se pueden modificar** despu√©s de creadas: no puedes agregar, eliminar ni cambiar elementos.
- Se usan cuando los datos **no deben cambiar**: coordenadas, d√≠as de la semana, configuraciones fijas.

### Crear una tupla

````python
colores = "rojo","verde","azul"
print(type(colores))

print(colores[0])  
print(colores[1])
print(colores[2])

````

### Acceder a elementos

Se accede **igual que en las listas**, con √≠ndices.

````python
colores = ("rojo", "verde", "azul")

print(colores[0])   # rojo
print(colores[-1])  # azul
print(len(colores)) # 3
````

### üö´ Inmutabilidad ‚Äî No se pueden modificar

Si intentas cambiar un elemento de una tupla, Python lanza un **error**.

````python
colores = ("rojo", "verde", "azul")

print("Las tuplas son inmutables: no se pueden cambiar despu√©s de crearse.")
````

### ¬øCu√°ndo usar tuplas?

Usa tuplas cuando los datos **no deben cambiar** nunca:

- **Coordenadas geogr√°ficas** (latitud, longitud)
- **Colores RGB** (255, 0, 128)
- **D√≠as de la semana** ("lunes", "martes", ...)
- **Datos de configuraci√≥n fijos**

#### üìç Ejemplo ‚Äî Coordenadas de ciudades

````python
# Coordenadas (latitud, longitud)
lima = (-12.0464, -77.0428)
bogota = (4.7110, -74.0721)
madrid = (40.4168, -3.7038)

print(f"Lima est√° en latitud {lima[0]} y longitud {lima[1]}")
print(f"Bogot√° est√° en latitud {bogota[0]} y longitud {bogota[1]}")
print(f"Madrid est√° en latitud {madrid[0]} y longitud {madrid[1]}")

# Desempaquetado de tuplas
lat, lon = lima
print(f"\nDesempaquetado ‚Üí Latitud: {lat}, Longitud: {lon}")
````

#### üé® Ejemplo ‚Äî Colores RGB

````python
rojo = (255, 0, 0)
verde = (0, 255, 0)
azul = (0, 0, 255)
blanco = (255, 255, 255)

colores = [rojo, verde, azul, blanco]

for color in colores:
    roja, verde, azul = color  # Desempaquetado
    print(f"R={roja}, G={verde}, B={azul}")
````

### Tupla vs Lista ‚Äî Resumen

| Caracter√≠stica | Lista `[]` | Tupla `()` |
| --- | --- | --- |
| ¬øMutable? | ‚úÖ S√≠ | ‚ùå No |
| ¬øSe puede agregar? | ‚úÖ `append()` | ‚ùå No |
| ¬øSe puede eliminar? | ‚úÖ `remove()` | ‚ùå No |
| ¬øSe puede ordenar? | ‚úÖ `sort()` | ‚ùå No |
| ¬øPara qu√© sirve? | Datos que cambian | Datos fijos |

## üìñ Diccionarios

### Concepto

Un **diccionario** es una colecci√≥n de **pares clave-valor**. En vez de acceder por posici√≥n (como en listas), accedes por una **clave** (nombre, etiqueta).

- Se crean con **llaves** `{}`.
- Cada par se escribe como `clave: valor`.
- Las claves son **√∫nicas** (no se repiten).
- Los valores pueden ser de **cualquier tipo**.

Piensa en un diccionario real: buscas una **palabra** (clave) y encuentras su **definici√≥n** (valor).

### Crear un diccionario

````python
# Diccionario de una persona
persona = {
    "nombre": "Ana",
    "edad": 25,
    "ciudad": "Lima"
}
print(persona)
print("Nombre:", persona["nombre"])
print("Edad:", persona["edad"])
print("Ciudad:", persona["ciudad"])
````

### Acceder a valores

Se accede usando la **clave** entre corchetes `[]` o con el m√©todo `.get()`.

````python
persona = {"nombre": "Ana", "edad": 25, "ciudad": "Lima"}

# Con corchetes
print(persona["nombre"])  # Ana
print(persona.get("edad",30))    # 25

print(persona.get("fecha_nacimiento", "No registrado"))  # Valor por defecto
````

### Agregar y modificar elementos

- Si la clave **no existe**, se **agrega**.
- Si la clave **ya existe**, se **modifica** el valor.

````python
persona = {"nombre": "Ana", "edad": 25}

# Agregar nueva clave
persona["email"] = "ana@email.com"
persona["telefono"] = "123456789"
print("Despu√©s de agregar email:", persona)
````

### Recorrer un diccionario

Puedes recorrer las **claves**, los **valores** o **ambos**.

````python
persona = {"nombre": "Ana", "edad": 25, "ciudad": "Lima"}

# Recorrer claves
print("--- Claves ---")
for clave in persona:
    print(clave)
    

# Recorrer valores
print("\n--- Valores ---")
for valor in persona.values():
    print(valor)
    
    

# Recorrer claves y valores juntos
print("\n--- Claves y valores ---")
for clave, valor in persona.items():
    print(f"{clave}: {valor}")
````

#### üì± Ejemplo ‚Äî Agenda de contactos

````python
agenda = {
    "Ana": "999-111-222",
    "Luis": "999-333-444",
    "Carlos": "999-555-666"
}
nombre = "Luis"
# Buscar un contacto
if nombre in agenda:
    print(f"üìû {nombre}: {agenda[nombre]}")
else:
    print(f"‚ùå {nombre} no est√° en la agenda")
````

#### üí∞ Ejemplo ‚Äî Gastos por categor√≠a

````python
gastos = {
    "comida": 350,
    "transporte": 120,
    "entretenimiento": 80,
    "ropa": 200,
    "servicios": 150
}
for categoria,valor in gastos.items():
    print(f"{categoria}: ${valor}")
````

### M√©todos √∫tiles de diccionarios

| M√©todo | Descripci√≥n |
| --- | --- |
| `dict.keys()` | Devuelve todas las claves |
| `dict.values()` | Devuelve todos los valores |
| `dict.items()` | Devuelve pares (clave, valor) |
| `dict.get(clave)` | Devuelve el valor o `None` si no existe |
| `dict.pop(clave)` | Elimina y devuelve el valor de esa clave |
| `dict.update(otro_dict)` | Actualiza con otro diccionario |
| `len(dict)` | Cantidad de pares clave-valor |

````python
print(gastos.keys())
print(gastos.values())
print(gastos.items())

print(len(gastos))
print("comida:", gastos.get("comida"))
````

````python
notas = {"mate": 15, "historia": 12, "ciencias": 18}

print("Claves:", list(notas.keys()))
print("Valores:", list(notas.values()))
print("Pares:", list(notas.items()))
print("Cantidad de cursos:", len(notas))

# pop: sacar una clave
nota_historia = notas.pop("historia")
print(f"Se sac√≥ historia con nota {nota_historia}")
print("Diccionario ahora:", notas)

# update: fusionar diccionarios
notas.update({"arte": 16, "ingl√©s": 14})
print("Despu√©s de update:", notas)
````

## üîµ Sets (Conjuntos)

### Concepto

Un **set** (conjunto) es una colecci√≥n **desordenada** de elementos **√∫nicos** (sin duplicados).

- Se crean con **llaves** `{}` o con `set()`.
- **No permite elementos repetidos** ‚Üí si agregas uno que ya existe, lo ignora.
- **No tiene orden** ‚Üí no puedes acceder por √≠ndice.
- Muy √∫til para **eliminar duplicados** y hacer **operaciones de conjuntos** (uni√≥n, intersecci√≥n, etc.).

### Crear un set

````python
# Set de frutas
frutas = {"manzana", "pl√°tano", "fresa", "uva","uva","uva","uva"}
print(frutas)
print(type(frutas))
````

### üßπ Eliminar duplicados de una lista

Uno de los usos m√°s comunes de los sets: **convertir una lista con duplicados en una sin duplicados**.

````python
# Lista con nombres repetidos
inscritos = ["Ana", "Luis", "Ana", "Carlos", "Luis", "Ana", "Beatriz"]
print("Lista original:", inscritos)
print(f"Total: {len(inscritos)} nombres")

inscritos=list(set(inscritos))
print("Sin duplicados:", inscritos)
print(f"Total √∫nicos: {len(inscritos)} nombres")
````

### Agregar y eliminar elementos

````python
frutas = {"manzana", "pl√°tano", "fresa"}
print("Frutas:", frutas)
frutas.add("uva")
print("Despu√©s de agregar 'uva':", frutas)
frutas.add("uva")
print("Despu√©s de agregar 'uva' otra vez (sin cambio):", frutas)
````

### Operaciones de conjuntos

Los sets soportan las mismas operaciones que los conjuntos en matem√°ticas.

| Operaci√≥n | M√©todo | Operador | Resultado |
| --- | --- | --- | --- |
| Uni√≥n | `a.union(b)` | `a \| b` | Todos los elementos de ambos |
| Intersecci√≥n | `a.intersection(b)` | `a & b` | Solo los que est√°n en ambos |
| Diferencia | `a.difference(b)` | `a - b` | Los que est√°n en `a` pero no en `b` |
| Diferencia sim√©trica | `a.symmetric_difference(b)` | `a ^ b` | Los que est√°n en uno u otro, pero no en ambos |

````python
# Ejemplo: materias de dos estudiantes
materias_ana = {"mate", "historia", "ciencias", "arte"}
materias_luis = {"mate", "ingl√©s", "ciencias", "deportes"}

# Uni√≥n: todas las materias entre los dos
print("üìö Todas las materias:", materias_ana | materias_luis)

print("Los cursos de ambos estudiantes:", materias_ana & materias_luis)

print("Los cursos que solo tiene Ana:", materias_ana - materias_luis)

print("Los cursos que no comparten:", materias_ana ^ materias_luis)
````

#### üéüÔ∏è Ejemplo ‚Äî Asistentes a eventos

````python
evento_viernes = {"Ana", "Luis", "Carlos", "Mar√≠a"}
evento_sabado = {"Luis", "Beatriz", "Carlos", "Diego"}


# ¬øQui√©n fue a ambos eventos?
print(evento_viernes.union(evento_sabado))

# ¬øQui√©n fue al menos a uno?
evento_viernes.symmetric_difference(evento_sabado)
````

```python
fiesta1 = ["Ana", "Luis", "Carlos", "Mar√≠a", "Ana"]
```

Usando sets:
1. Muestra los asistentes √∫nicos de cada fiesta.

3. ¬øQui√©n fue **solo** a la fiesta 1?
