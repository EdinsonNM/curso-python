## ğŸ”§ DefiniciÃ³n de funciones

### Concepto

Una **funciÃ³n** es un bloque de cÃ³digo con un **nombre** que realiza una tarea especÃ­fica. Se define una vez y se puede **llamar** (ejecutar) cuantas veces se necesite.

### Sintaxis

````python
def nombre_de_la_funcion():
    # cÃ³digo que ejecuta la funciÃ³n
    pass
````

- **`def`**: palabra reservada que indica que estÃ¡s **definiendo** una funciÃ³n.
- **`nombre_de_la_funcion`**: el nombre que le das (sigue las mismas reglas que las variables).
- **`()`**: parÃ©ntesis donde irÃ¡n los parÃ¡metros (si los hay).
- **`:`**: dos puntos obligatorios al final de la lÃ­nea.
- **Bloque indentado**: el cÃ³digo que se ejecuta cuando la funciÃ³n es llamada.

ğŸš¨ **Regla clave**: Definir una funciÃ³n **no la ejecuta**. Solo la guarda en memoria. Para ejecutarla, hay que **llamarla** por su nombre.

### Ejemplo bÃ¡sico â€” Definir y llamar

````python
# Definir la funciÃ³n
def saludar():
    print("Â¡Hola! Bienvenido al curso de Python ğŸ")

# Llamar la funciÃ³n
saludar()
saludar()  # Se puede llamar cuantas veces quieras
saludar()
````

## ğŸ“¨ ParÃ¡metros

### Concepto

Los **parÃ¡metros** son variables que van dentro de los parÃ©ntesis de la funciÃ³n. Permiten que la funciÃ³n reciba **datos de entrada** para trabajar con ellos.

- **ParÃ¡metro**: la variable en la definiciÃ³n de la funciÃ³n.
- **Argumento**: el valor que le pasas cuando la llamas.

Es como un molde de galletas: el **parÃ¡metro** es la forma del molde, y el **argumento** es la masa que le pones cada vez.

### FunciÃ³n con un parÃ¡metro

````python
def saludar(nombre):  # 'nombre' es el parÃ¡metro
    print(f"Â¡Hola, {nombre}! ğŸ‘‹")

# 'Ana', 'Luis', 'Carlos' son los argumentos
saludar("Ana")
saludar("Luis")
saludar("Carlos")
````

## ğŸ“¤ Retorno de valores (`return`)

### Concepto

Hasta ahora nuestras funciones solo **imprimen** resultados. Pero en la mayorÃ­a de casos, lo que necesitas es que la funciÃ³n **devuelva** un valor para que puedas **usarlo despuÃ©s** en otra parte del cÃ³digo.

Para eso se usa **`return`**.

- `return` **devuelve** un valor al lugar donde se llamÃ³ la funciÃ³n.
- La funciÃ³n **termina inmediatamente** cuando llega a `return`.
- El valor devuelto se puede **guardar en una variable**, usar en una operaciÃ³n, imprimir, etc.

### Diferencia: `print` vs `return`

| `print()` | `return` |
| --- | --- |
| Muestra algo en pantalla | Devuelve un valor al cÃ³digo |
| No se puede guardar el resultado | Se puede guardar en una variable |
| Solo para "ver" | Para **usar** el resultado |

````python
# Con print: solo muestra, no puedes usar el resultado
def sumar_print(a, b):
    print(a + b)

resultado = sumar_print(3, 5)  # Imprime 8
print(f"Resultado guardado: {resultado}")  # None (no devolviÃ³ nada)
````

````python
# Con return: devuelve el valor, puedes usarlo
def sumar_return(a, b):
    return a + b

resultado = sumar_return(3, 5)  # No imprime nada, pero guarda el 8
print(f"Resultado guardado: {resultado}")  # 8

# Ahora puedes usarlo en operaciones
doble = resultado * 2
print(f"El doble es: {doble}")  # 16
````

### Ejemplo bÃ¡sico con `return`

````python
def cuadrado(numero):
    return numero ** 2

resultado = cuadrado(5)
print(f"El cuadrado de 5 es {resultado}")  # 25

# TambiÃ©n puedes usarlo directamente
print(f"El cuadrado de 9 es {cuadrado(9)}")  # 81
````

### Retornar mÃºltiples valores

Python permite devolver **varios valores** separados por comas. En realidad devuelve una **tupla**.

````python
def operaciones(a, b):
    suma = a + b
    resta = a - b
    multiplicacion = a * b
    return suma, resta, multiplicacion

# Desempaquetado
s, r, m = operaciones(10, 3)
print(f"Suma: {s}")            # 13
print(f"Resta: {r}")           # 7
print(f"MultiplicaciÃ³n: {m}")  # 30

# TambiÃ©n puedes guardarlo como tupla
resultado = operaciones(10, 3)
print(f"Resultado completo: {resultado}")  # (13, 7, 30)
````

### `return` termina la funciÃ³n

Todo lo que estÃ© **despuÃ©s** de `return` dentro de la funciÃ³n **no se ejecuta**.

````python
def verificar_edad(edad):
    if edad >= 18:
        return "Mayor de edad âœ…"
    return "Menor de edad âŒ"
    # Esto nunca se ejecuta:
    print("Esta lÃ­nea no se ejecuta nunca")

print(verificar_edad(25))  # Mayor de edad âœ…
print(verificar_edad(15))  # Menor de edad âŒ
````

## ğŸŒ Variables locales y globales

### Concepto

No todas las variables son iguales. En Python, el lugar donde **creas** una variable determina dÃ³nde se puede **usar**. A esto se le llama **alcance** (scope).

| Tipo | DÃ³nde se crea | DÃ³nde se puede usar |
| --- | --- | --- |
| **Variable local** | Dentro de una funciÃ³n | Solo dentro de esa funciÃ³n |
| **Variable global** | Fuera de cualquier funciÃ³n | En todo el programa |

### Variables locales

Las variables creadas **dentro** de una funciÃ³n solo existen **dentro** de esa funciÃ³n. Cuando la funciÃ³n termina, la variable **desaparece**.

````python
def mi_funcion():
    mensaje = "Soy una variable local"  # Variable LOCAL
    print(mensaje)

mi_funcion()  # Funciona: "Soy una variable local"

# Esto darÃ­a error si se descomenta:
# print(mensaje)  # âŒ NameError: 'mensaje' is not defined
````

### Variables globales

Las variables creadas **fuera** de cualquier funciÃ³n son **globales** y se pueden **leer** desde cualquier lugar del programa.

````python
saludo = "Â¡Hola, mundo!"  # Variable GLOBAL

def mostrar_saludo():
    print(saludo)  # Puede LEER la variable global

mostrar_saludo()   # Â¡Hola, mundo!
print(saludo)      # Â¡Hola, mundo! (tambiÃ©n funciona fuera)
````

### âš ï¸ Problema: modificar una variable global desde una funciÃ³n

Si intentas **modificar** una variable global dentro de una funciÃ³n, Python crea una **nueva variable local** con el mismo nombre. La global no cambia.

````python
contador = 0  # Variable global

def incrementar():
    # Esto crea una variable LOCAL llamada 'contador', no modifica la global
    contador = 10
    print(f"Dentro de la funciÃ³n: {contador}")  # 10

incrementar()
print(f"Fuera de la funciÃ³n: {contador}")  # 0 (no cambiÃ³)
````

### La palabra `global`

Si realmente necesitas **modificar** una variable global desde dentro de una funciÃ³n, usas la palabra reservada `global`.

âš ï¸ **Pero es una mala prÃ¡ctica en la mayorÃ­a de casos.** Es mejor usar `return` para devolver valores.

````python
contador = 0

def incrementar():
    global contador  # Ahora sÃ­ modifica la variable global
    contador += 1
    print(f"Dentro: {contador}")

incrementar()  # Dentro: 1
incrementar()  # Dentro: 2
incrementar()  # Dentro: 3
print(f"Fuera: {contador}")  # Fuera: 3
````

### âœ… Mejor enfoque: usar `return` en lugar de `global`

````python
def incrementar(contador):
    contador += 1
    return contador

mi_contador = 0
mi_contador = incrementar(mi_contador)  # 1
mi_contador = incrementar(mi_contador)  # 2
mi_contador = incrementar(mi_contador)  # 3
print(f"Contador final: {mi_contador}")  # 3

# Mucho mÃ¡s claro y seguro que usar 'global'
````

## ğŸ“ Buenas prÃ¡cticas en funciones

### 1. Usa nombres descriptivos

El nombre de la funciÃ³n debe decir **quÃ© hace**, no cÃ³mo lo hace.

````python
# âŒ Mal: nombre poco claro
def f(x):
    return x * 1.18

# âœ… Bien: nombre descriptivo
def calcular_precio_con_igv(precio):
    return precio * 1.18

print(calcular_precio_con_igv(100))  # 118.0
````

### 2. Una funciÃ³n = una tarea

Cada funciÃ³n debe hacer **una sola cosa** y hacerla bien.

````python
# âŒ Mal: hace demasiadas cosas
def procesar_todo(notas):
    promedio = sum(notas) / len(notas)
    if promedio >= 11:
        print("Aprobado")
    else:
        print("Desaprobado")
    print(f"La nota mÃ¡s alta es {max(notas)}")
    print(f"La nota mÃ¡s baja es {min(notas)}")

# âœ… Bien: cada funciÃ³n hace una sola cosa
def calcular_promedio(notas):
    return sum(notas) / len(notas)

def obtener_estado(promedio):
    if promedio >= 11:
        return "Aprobado"
    return "Desaprobado"

def nota_maxima(notas):
    return max(notas)

def nota_minima(notas):
    return min(notas)

# Uso
notas = [15, 12, 18, 9, 14]
prom = calcular_promedio(notas)
print(f"Promedio: {prom:.1f} â†’ {obtener_estado(prom)}")
print(f"MÃ¡xima: {nota_maxima(notas)} | MÃ­nima: {nota_minima(notas)}")
````

### 3. Usa `return` en lugar de `print` (cuando sea posible)

Las funciones que **devuelven** valores son mÃ¡s **flexibles** y **reutilizables** que las que solo imprimen.

````python
# âŒ Limitado: solo imprime, no puedes usar el resultado
def area_circulo_print(radio):
    print(3.1416 * radio ** 2)

# âœ… Flexible: devuelve el valor para usarlo en cualquier lado
def area_circulo(radio):
    return 3.1416 * radio ** 2

# Ahora puedes usar el resultado
area = area_circulo(5)
print(f"Ãrea: {area:.2f}")

# O compararlo
if area_circulo(10) > 200:
    print("El cÃ­rculo es grande")
````

### 4. Evita usar `global`

Pasa los datos como **parÃ¡metros** y devuelve resultados con **`return`**.

````python
# âŒ Mal: depende de una variable global
saldo = 100

def retirar_mal(monto):
    global saldo
    saldo -= monto

# âœ… Bien: todo se pasa y devuelve explÃ­citamente
def retirar(saldo, monto):
    if monto > saldo:
        print("Saldo insuficiente")
        return saldo
    return saldo - monto

mi_saldo = 100
mi_saldo = retirar(mi_saldo, 30)
print(f"Saldo: S/{mi_saldo}")  # 70

mi_saldo = retirar(mi_saldo, 50)
print(f"Saldo: S/{mi_saldo}")  # 20
````

### 5. Documenta tus funciones con docstrings

Un **docstring** es un texto entre triples comillas que explica quÃ© hace la funciÃ³n. Se coloca justo despuÃ©s del `def`.

````python
def calcular_igv(precio, tasa=18):
    """Calcula el precio con IGV incluido.
    
    ParÃ¡metros:
        precio: El precio base del producto.
        tasa: El porcentaje de IGV (por defecto 18%).
    
    Retorna:
        El precio final con IGV incluido.
    """
    return precio * (1 + tasa / 100)

# Puedes ver la documentaciÃ³n con help()
help(calcular_igv)

print(f"\nPrecio final: S/{calcular_igv(100):.2f}")
````
