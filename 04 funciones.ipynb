## üîß Definici√≥n de funciones

### Concepto

Una **funci√≥n** es un bloque de c√≥digo con un **nombre** que realiza una tarea espec√≠fica. Se define una vez y se puede **llamar** (ejecutar) cuantas veces se necesite.

### Sintaxis

````python
def nombre_de_la_funcion():
    # c√≥digo que ejecuta la funci√≥n
    pass
````

````python
# aqui defino como ser√° la funci√≥n
def saludar_por_semana_santa():
    print("¬°Feliz Semana Santa!")

# aqu√≠ ejecuto la funci√≥n
saludar_por_semana_santa()
saludar_por_semana_santa()
saludar_por_semana_santa()
````

- **`def`**: palabra reservada que indica que est√°s **definiendo** una funci√≥n.
- **`nombre_de_la_funcion`**: el nombre que le das (sigue las mismas reglas que las variables).
- **`()`**: par√©ntesis donde ir√°n los par√°metros (si los hay).
- **`:`**: dos puntos obligatorios al final de la l√≠nea.
- **Bloque indentado**: el c√≥digo que se ejecuta cuando la funci√≥n es llamada.

üö® **Regla clave**: Definir una funci√≥n **no la ejecuta**. Solo la guarda en memoria. Para ejecutarla, hay que **llamarla** por su nombre.

### Ejemplo b√°sico ‚Äî Definir y llamar

````python
# Definir la funci√≥n
def saludar():
    print("¬°Hola! Bienvenido al curso de Python üêç")

# Llamar la funci√≥n
saludar()
saludar()  # Se puede llamar cuantas veces quieras
saludar()
saludar()
````

## üì§ Retorno de valores (`return`)

### Concepto

Hasta ahora nuestras funciones solo **imprimen** resultados. Pero en la mayor√≠a de casos, lo que necesitas es que la funci√≥n **devuelva** un valor para que puedas **usarlo despu√©s** en otra parte del c√≥digo.

Para eso se usa **`return`**.

- `return` **devuelve** un valor al lugar donde se llam√≥ la funci√≥n.
- La funci√≥n **termina inmediatamente** cuando llega a `return`.
- El valor devuelto se puede **guardar en una variable**, usar en una operaci√≥n, imprimir, etc.

### Diferencia: `print` vs `return`

| `print()` | `return` |
| --- | --- |
| Muestra algo en pantalla | Devuelve un valor al c√≥digo |
| No se puede guardar el resultado | Se puede guardar en una variable |
| Solo para "ver" | Para **usar** el resultado |

````python
import random
def generar_suerte():
    return random.randint(1, 100)

resultado=generar_suerte()
print(f"Tu n√∫mero de suerte es: {resultado}")
if(resultado > 50):
    print("Eres muy afortunado hoy! üéâ")
else:
    print("Hoy no es tu d√≠a de suerte, ¬°pero no te desanimes! üçÄ")
````

````python
# Con return: devuelve el valor, puedes usarlo
def sumar_return(a, b):
    return a + b

resultado = sumar_return(3, 5)  # No imprime nada, pero guarda el 8
print(f"Resultado guardado: {resultado}")  # 8

# Ahora puedes usarlo en operaciones
doble = resultado * 2
print(f"El doble es: {doble}")  # 16
````

### Ejemplo b√°sico con `return`

### Retornar m√∫ltiples valores

Python permite devolver **varios valores** separados por comas. En realidad devuelve una **tupla**.

````python
def operaciones(a, b):
    suma = a + b
    resta = a - b
    multiplicacion = a * b
    return suma, resta, multiplicacion

# Desempaquetado
s, r, m = operaciones(10, 3)
print(f"Suma: {s}")            # 13
print(f"Resta: {r}")           # 7
print(f"Multiplicaci√≥n: {m}")  # 30

resultado = operaciones(10, 5)
print(f"Resultado completo: {resultado}")  # (15, 5, 50)
print("suma:", resultado[0])  # 15
print("resta:", resultado[1])  # 5
print("multiplicaci√≥n:", resultado[2])  # 50
````

## üåê Variables locales y globales

### Concepto

No todas las variables son iguales. En Python, el lugar donde **creas** una variable determina d√≥nde se puede **usar**. A esto se le llama **alcance** (scope).

| Tipo | D√≥nde se crea | D√≥nde se puede usar |
| --- | --- | --- |
| **Variable local** | Dentro de una funci√≥n | Solo dentro de esa funci√≥n |
| **Variable global** | Fuera de cualquier funci√≥n | En todo el programa |

### Variables locales

Las variables creadas **dentro** de una funci√≥n solo existen **dentro** de esa funci√≥n. Cuando la funci√≥n termina, la variable **desaparece**.

````python
def mi_funcion():
    mensaje = "Soy una variable local"  # Variable LOCAL
    print(mensaje)

mi_funcion()  # Funciona: "Soy una variable local"
print(mensaje)
````

### Variables globales

Las variables creadas **fuera** de cualquier funci√≥n son **globales** y se pueden **leer** desde cualquier lugar del programa.

````python
saludo = "¬°Hola, mundo!"  # Variable GLOBAL

def mostrar_saludo():
    saludo="Hola Python"
    print(saludo)  # Puede LEER la variable global

mostrar_saludo()   
print(saludo)     
````

### ‚ö†Ô∏è Problema: modificar una variable global desde una funci√≥n

Si intentas **modificar** una variable global dentro de una funci√≥n, Python crea una **nueva variable local** con el mismo nombre. La global no cambia.

````python
contador = 0  # Variable global

def incrementar():
    # Esto crea una variable LOCAL llamada 'contador', no modifica la global
    contador = 10
    print(f"Dentro de la funci√≥n: {contador}")  # 10

incrementar()
print(f"Fuera de la funci√≥n: {contador}")  # 0 (no cambi√≥)
````

### La palabra `global`

Si realmente necesitas **modificar** una variable global desde dentro de una funci√≥n, usas la palabra reservada `global`.

‚ö†Ô∏è **Pero es una mala pr√°ctica en la mayor√≠a de casos.** Es mejor usar `return` para devolver valores.

````python
contador = 0

def incrementar():
    global contador  # Ahora s√≠ modifica la variable global
    contador += 1
    print(f"Dentro: {contador}")

incrementar()  # Dentro: 1
incrementar()  # Dentro: 2
incrementar()  # Dentro: 3
print(f"Fuera: {contador}")  # Fuera: 3
````

### ‚úÖ Mejor enfoque: usar `return` en lugar de `global`

````python
def incrementar(contador):
    contador += 1
    return contador

mi_contador = 0
mi_contador = incrementar(mi_contador)  # 1
mi_contador = incrementar(mi_contador)  # 2
mi_contador = incrementar(mi_contador)  # 3
print(f"Contador final: {mi_contador}")  # 3

# Mucho m√°s claro y seguro que usar 'global'
````

## üìè Buenas pr√°cticas en funciones

### 1. Usa nombres descriptivos

El nombre de la funci√≥n debe decir **qu√© hace**, no c√≥mo lo hace.

````python
# ‚ùå Mal: nombre poco claro
def f(x):
    return x * 1.18

# ‚úÖ Bien: nombre descriptivo
def calcular_precio_con_igv(precio):
    return precio * 1.18

print(calcular_precio_con_igv(100))  # 118.0
````

### 2. Una funci√≥n = una tarea

Cada funci√≥n debe hacer **una sola cosa** y hacerla bien.

````python
# ‚ùå Mal: hace demasiadas cosas
def procesar_todo(notas):
    promedio = sum(notas) / len(notas)
    if promedio >= 11:
        print("Aprobado")
    else:
        print("Desaprobado")
    print(f"La nota m√°s alta es {max(notas)}")
    print(f"La nota m√°s baja es {min(notas)}")

def ejecutar_proceso(notas):
    promedio=calcular_promedio(notas)
    obteniendo_estado=obtener_estado(promedio)
    nota_max=nota_maxima(notas)
    nota_min=nota_minima(notas)


# ‚úÖ Bien: cada funci√≥n hace una sola cosa
def calcular_promedio(notas):
    return sum(notas) / len(notas)

def obtener_estado(promedio):
    if promedio >= 11:
        return "Aprobado"
    return "Desaprobado"

def nota_maxima(notas):
    return max(notas)

def nota_minima(notas):
    return min(notas)




# Uso
notas = [15, 12, 18, 9, 14]
prom = calcular_promedio(notas)
print(f"Promedio: {prom:.1f} ‚Üí {obtener_estado(prom)}")
print(f"M√°xima: {nota_maxima(notas)} | M√≠nima: {nota_minima(notas)}")
````

### 3. Usa `return` en lugar de `print` (cuando sea posible)

Las funciones que **devuelven** valores son m√°s **flexibles** y **reutilizables** que las que solo imprimen.

````python
# ‚ùå Limitado: solo imprime, no puedes usar el resultado
def area_circulo_print(radio):
    print(3.1416 * radio ** 2)

# ‚úÖ Flexible: devuelve el valor para usarlo en cualquier lado
def area_circulo(radio):
    return 3.1416 * radio ** 2

# Ahora puedes usar el resultado
area = area_circulo(5)
print(f"√Årea: {area:.2f}")

# O compararlo
if area_circulo(10) > 200:
    print("El c√≠rculo es grande")
````

### 4. Evita usar `global`

Pasa los datos como **par√°metros** y devuelve resultados con **`return`**.

````python
# ‚ùå Mal: depende de una variable global
saldo = 100

def retirar_mal(monto):
    global saldo
    saldo -= monto

# ‚úÖ Bien: todo se pasa y devuelve expl√≠citamente
def retirar(saldo, monto):
    if monto > saldo:
        print("Saldo insuficiente")
        return saldo
    return saldo - monto

mi_saldo = 100
mi_saldo = retirar(mi_saldo, 30)
print(f"Saldo: S/{mi_saldo}")  # 70

mi_saldo = retirar(mi_saldo, 50)
print(f"Saldo: S/{mi_saldo}")  # 20
````

### 5. Documenta tus funciones con docstrings

Un **docstring** es un texto entre triples comillas que explica qu√© hace la funci√≥n. Se coloca justo despu√©s del `def`.

````python
def calcular_igv(precio, tasa=18):
    """Calcula el igv
    \nParametros:
        precio: El precio base del producto.
        tasa: El porcentaje de IGV (por defecto 18%).
     Retorna:
        El precio final con IGV incluido.
    """
    
    return precio * (1 + tasa / 100)

# Puedes ver la documentaci√≥n con help()
help(calcular_igv)

print(f"\nPrecio final: S/{calcular_igv(100):.2f}")
````

#### üå°Ô∏è Ejercicio 6: El convertidor de temperatura
üü° Nivel intermedio

Crea dos funciones:
- `celsius_a_fahrenheit(c)` ‚Üí devuelve `c * 9/5 + 32`
- `fahrenheit_a_celsius(f)` ‚Üí devuelve `(f - 32) * 5/9`

El usuario elige la conversi√≥n y el programa muestra el resultado.

_(Para cuando no sabes si 72¬∞F es fr√≠o o calor.)_

````python
def celsius_a_fahrenheit(celsius):
    return celsius * 9/5 + 32

def farenheit_a_celsius(fahrenheit):
    return (fahrenheit - 32) * 5/9

def es_farenheit(tipo):
    return tipo.upper() == "F"

def es_celsius(tipo):
    return tipo.upper() == "C"

# c C F f upper, lower
temperatura = float(input("qu√© temperatura quieres convertir: "))
tipo_conversion=input("a qu√© unidad quieres convertir (C/F): ")

if es_farenheit(tipo_conversion):
    resultado = celsius_a_fahrenheit(temperatura)
    print(f"{temperatura}¬∞C son {resultado:.2f}¬∞F")
elif es_celsius(tipo_conversion):
    resultado = farenheit_a_celsius(temperatura)
    print(f"{temperatura}¬∞F son {resultado:.2f}¬∞C")
else:
    print("Unidad de conversi√≥n no v√°lida.")
````

#### üîê Ejercicio 7: El validador de contrase√±as

Crea una funci√≥n `validar(password)` que devuelva `True` si la contrase√±a cumple:
- Tiene al menos 8 caracteres.
- Tiene al menos un n√∫mero.

Pide la contrase√±a al usuario y muestra si es v√°lida o no, explicando qu√© falta.

_("123" no es contrase√±a. Ni "password". Ni tu nombre.)_

- `nota_mas_alta(notas)` ‚Üí devuelve la mayor.
- `nota_mas_baja(notas)` ‚Üí devuelve la menor.
- `estado(promedio)` ‚Üí devuelve "Aprobado" si >= 11, "Desaprobado" si no.

Pide 5 notas al usuario y muestra el reporte completo.

_(Tu bolet√≠n, pero hecho en Python.)_

````python
def promedio(notas):
    return sum(notas) / len(notas)

def nota_mas_alta(notas):
    return max(notas)

def nota_mas_baja(notas):
    return min(notas)

def estado(promedio):
    if promedio >= 11:
        return "Aprobado"
    return "Desaprobado"

notas=[]
while len(notas) < 5:
    nota=int(input(f"ingresa la nota: {len(notas)+1}"))
    if nota>0 and nota<=20:
        notas.append(nota)


promedio=promedio(notas)    
print("reporte de notas:")
print(promedio)
print(nota_mas_alta(notas))
print(nota_mas_baja(notas))
print(estado(promedio))
````
