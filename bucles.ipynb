## Bucles
 Las estructuras de iteraci√≥n permiten repetir secciones de c√≥digo. A menudo se denominan bucles porque permiten al programa volver a una l√≠nea ya ejecutada. El proceso de ejecutar una repetici√≥n de un bucle tambi√©n se conoce como iteraci√≥n del bucle.
### WHILE
Un bucle while permite repetir un bloque de c√≥digo mientras una condici√≥n sea verdadera.

```python
while True:
    number = int(input("Please type in a number, -1 to quit: "))

    if number == -1:
        break

    print(number ** 2)

print("Thanks and bye!")
```

Todo while necesita una forma de terminar.

Si la condici√≥n nunca cambia ‚Üí se crea un bucle infinito.

Ejemplo peligroso:

```python

while True:
    print("Hola")
```

Esto nunca se detiene.
