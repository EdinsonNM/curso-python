## Random

El mÃ³dulo random es parte de la biblioteca estÃ¡ndar de Python (viene incluido, no hay que instalar nada).

Su propÃ³sito principal es generar nÃºmeros pseudoaleatorios y hacer selecciones al azar de forma sencilla y multiplataforma.


Funciones mÃ¡s usadas:
- random.random() â†’ float entre 0.0 y 1.0 (excluyendo 1.0)
- random.randint(a, b) â†’ entero entre a y b (incluyendo ambos)
- random.randrange(start, stop, step) â†’ similar a range() pero aleatorio
- random.choice(secuencia) â†’ elige 1 elemento al azar
- random.choices(secuencia, weights=None, k=1) â†’ elige con posibles repeticiones y pesos
- random.sample(secuencia, k) â†’ elige k elementos sin repetir
- random.shuffle(lista) â†’ desordena la lista en el lugar (in-place)


Ejemplo mÃ­nimo:

````python
import random

print(random.random())
print(random.randint(1, 100))

frutas=["manzana", "banana", "cereza", "naranja"]
print("La fruta elegida al azar es:", random.choice(frutas))
````

### Ejercicio: "Adivina el nÃºmero (computadora elige)"

Crea un programa que:

- Genere un nÃºmero secreto entre 1 y 50
- Pida al usuario que adivine (con input)
- Diga si el nÃºmero ingresado es mayor, menor o correcto
- Cuente los intentos hasta acertar
- Al final muestre cuÃ¡ntos intentos tomÃ³

````python
import random as r

secreto=r.randint(1,50)
contador=0
while True:  
    contador+=1
    numero_adivinado=int(input("adivina el nÃºmero secreto(1-50):"))
    if numero_adivinado > secreto:
        print("El nÃºmero adivinado es mayor")
    elif numero_adivinado < secreto:
        print("El nÃºmero adivinado es menor")
    else:
        print("Â¡Felicidades! Has adivinado el nÃºmero secreto.")
        break
    
print("NÃºmero de intentos:", contador)
````

````python
import random

secreto = random.randint(1, 50)
intentos = 0

print("Adivina el nÃºmero (estÃ¡ entre 1 y 50)")

while True:
    intento = int(input("Tu nÃºmero: "))
    intentos += 1
    
    if intento == secreto:
        print(f"Â¡Correcto! Lo lograste en {intentos} intento(s) ğŸ‰")
        break
    elif intento < secreto:
        print("MÃ¡s alto...")
    else:
        print("MÃ¡s bajo...")
````

## OpenAI

La librerÃ­a oficial openai es el cliente Python recomendado por OpenAI para interactuar con sus modelos (ChatGPT, GPT-4o, etc.), pero tambiÃ©n es compatible con otros proveedores que implementan la misma API (como Groq, Together, Fireworks, etc.).

Cuando usamos Groq, solo cambiamos dos cosas:

- la base_url â†’ apunta a los servidores de Groq
- la api_key â†’ es la clave que obtienes gratis en console.groq.com

Estructura tÃ­pica de una llamada:

````python
from openai import OpenAI

client = OpenAI(
    api_key="gsk_xxxxxxxxxxxxxxxxxxxxxxxx",
    base_url="https://api.groq.com/openai/v1"
)

respuesta = client.chat.completions.create(
    model="llama-3.1-70b-versatile",          # o llama-3.1-8b-instant, mixtral, etc.
    messages=[
        {"role": "system", "content": "Eres un profesor amable"},
        {"role": "user",   "content": "Explica quÃ© es una variable"}
    ],
    temperature=0.7,
    max_tokens=500
)

print(respuesta.choices[0].message.content)
````

````python
# chat_groq_while.py
# Requiere: pip install openai

import os
from openai import OpenAI

# =====================================
#  1. ConfiguraciÃ³n (cÃ¡mbialo por tu clave)
# =====================================
API_KEY = os.getenv("GROQ_API_KEY")   # mejor opciÃ³n: variable de entorno
# API_KEY = "gsk_xxxxxxxxxxxxxxxxxxxxxxxxxxxx"   # opciÃ³n temporal (no subas a git)

if not API_KEY:
    print("Error: No tienes GROQ_API_KEY configurada.")
    print("Ve a https://console.groq.com/keys â†’ crea una clave gratis")
    exit()

client = OpenAI(
    api_key=API_KEY,
    base_url="https://api.groq.com/openai/v1",
)

# =====================================
#  2. Personalidad del asistente (system prompt)
# =====================================
system_prompt = {
    "role": "system",
    "content": "Eres un asistente Ãºtil, directo y que responde en espaÃ±ol. Explicas cosas de programaciÃ³n de forma clara."
}

# Historial de mensajes (para que recuerde la conversaciÃ³n)
historial = [system_prompt]

print("Chat con Groq (Llama 3) - escribe 'salir' para terminar\n")
print("â”€" * 60)

while True:
    # 3. Pregunta del usuario
    mensaje = input("TÃº: ").strip()
    
    if mensaje.lower() in ["salir", "exit", "chau", "adiÃ³s", "q", "bye"]:
        print("\nÂ¡Hasta luego! ğŸ‘‹")
        break
    
    if not mensaje:
        print("Escribe algo...\n")
        continue
    
    # Agregamos el mensaje del usuario al historial
    historial.append({"role": "user", "content": mensaje})
    
    # 4. Llamada a la API
    try:
        respuesta = client.chat.completions.create(
            model="llama-3.1-8b-instant",          # rÃ¡pido y barato
            # model="llama-3.3-70b-versatile",     # mÃ¡s potente (cambia si quieres)
            messages=historial,
            temperature=0.7,
            max_tokens=600,
        )
        
        # 5. Obtenemos y mostramos la respuesta
        texto = respuesta.choices[0].message.content.strip()
        
        print("\nIA  :", texto)
        print("â”€" * 60)
        
        # Agregamos la respuesta al historial (para mantener contexto)
        historial.append({"role": "assistant", "content": texto})
        
    except Exception as e:
        print("\nError al conectar con Groq:", str(e))
        print("Verifica tu clave o conexiÃ³n a internet.\n")
````
