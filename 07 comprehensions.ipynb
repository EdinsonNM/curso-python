## 📘 Módulo 7: Python Comprehensions — Código más limpio y expresivo

Las **comprehensions** permiten crear listas, sets y diccionarios en una sola línea, de forma clara y compacta.

En este módulo aprenderás:
- List comprehensions
- Set comprehensions
- Dict comprehensions
- Comprehensions anidadas
- Generator expressions

---
## 🧾 1) List Comprehension

### Sintaxis

```python
[expresion for elemento in iterable]
```

````python
# Forma tradicional
cuadrados_tradicional = []
for n in range(1, 6):
    cuadrados_tradicional.append(n ** 2)

# Con list comprehension
cuadrados_comp = [n ** 2 for n in range(1, 6)]

print("Tradicional:", cuadrados_tradicional)
print("Comprehension:", cuadrados_comp)
````

````python
# Con filtro (if)
pares = [n for n in range(1, 21) if n % 2 == 0]
print("Números pares:", pares)
````

````python
# Con if-else dentro de la expresión
etiquetas = ["par" if n % 2 == 0 else "impar" for n in range(1, 11)]
print(etiquetas)
````

---
## 🔵 2) Set Comprehension

### Sintaxis
```python
{expresion for elemento in iterable}
```

````python
palabras = ["hola", "mundo", "hola", "python", "mundo"]
unicas_mayus = {p.upper() for p in palabras}
print("Original:", palabras)
print("Set en mayúsculas (sin duplicados):", unicas_mayus)
````

---
## 📖 3) Dict Comprehension

### Sintaxis
```python
{clave: valor for elemento in iterable}
```

````python
numeros = [1, 2, 3, 4, 5]
cuadrados = {n: n ** 2 for n in numeros}
print(cuadrados)
````

````python
# Dict comprehension con filtro
pares_cubo = {n: n ** 3 for n in range(1, 11) if n % 2 == 0}
print(pares_cubo)
````

---
## 🧩 4) Comprehensions anidadas

````python
matriz = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

plana = [valor for fila in matriz for valor in fila]
print("Matriz:", matriz)
print("Lista plana:", plana)
````

---
## ⚡ 5) Generator Expressions

Usan paréntesis `()` y generan valores bajo demanda (ahorran memoria).

````python
gen = (n ** 2 for n in range(1, 11))
print("Objeto generador:", gen)
print("Primeros 3 valores:", next(gen), next(gen), next(gen))
````

````python
# Suma eficiente con generador
total = sum(n for n in range(1, 1_000_001))
print("Suma del 1 al 1,000,000:", total)
````

## ✅ 6) Ejemplos prácticos

````python
# A) Precios con descuento
precios = [100, 250, 80, 120]
precios_descuento = [round(p * 0.9, 2) for p in precios]
print("Original:", precios)
print("Con descuento:", precios_descuento)
````

````python
# B) Filtrar emails válidos
emails = ["ana@email.com", "correo_invalido", "luis@mail.com", "hola@dominio"]
validos = [e for e in emails if "@" in e and "." in e.split("@")[-1]]
print("Emails válidos:", validos)
````

````python
# C) Longitud de palabras
palabras = ["python", "es", "genial"]
longitudes = {p: len(p) for p in palabras}
print(longitudes)
````

## ✏️ 7) Ejercicios

````python
# Ejercicio 1: Cuadrados del 1 al 20
cuadrados = [n ** 2 for n in range(1, 21)]
print(cuadrados)
````

````python
# Ejercicio 2: Múltiplos de 3
multiplos_3 = [n for n in range(1, 51) if n % 3 == 0]
print(multiplos_3)
````

````python
# Ejercicio 3: Par o impar
etiquetas = ["par" if n % 2 == 0 else "impar" for n in range(1, 16)]
print(etiquetas)
````

````python
# Ejercicio 4: Set sin duplicados
nombres = ["Ana", "Luis", "Ana", "Carlos", "Luis"]
nombres_unicos = {n.lower() for n in nombres}
print(nombres_unicos)
````

````python
# Ejercicio 5 y 6: Diccionarios por comprensión
cubos = {n: n ** 3 for n in range(1, 11)}
pares_cuadrado = {n: n ** 2 for n in range(1, 21) if n % 2 == 0}
print("Cubos:", cubos)
print("Pares y su cuadrado:", pares_cuadrado)
````

````python
# Ejercicio 7: Aplanar matriz
matriz = [[1, 2], [3, 4], [5, 6]]
plana = [x for fila in matriz for x in fila]
print(plana)
````

````python
# Ejercicio 8: Generador + sum
total = sum(n for n in range(1, 1_000_001))
print(total)
````

````python
# Ejercicio 9: Normalizar texto
frases = ["  Hola Mundo  ", " PYTHON es genial", "  Comprehensions  "]
normalizadas = [f.strip().lower() for f in frases]
print(normalizadas)
````

````python
# Ejercicio 10: Mini reporte de notas
notas = [10, 15, 18, 9, 20, 14]
aprobadas = [n for n in notas if n >= 11]
estado = {n: ("aprobado" if n >= 11 else "desaprobado") for n in notas}
categorias = {"aprobado" if n >= 11 else "desaprobado" for n in notas}

print("Aprobadas:", aprobadas)
print("Estado por nota:", estado)
print("Categorías únicas:", categorias)
````
